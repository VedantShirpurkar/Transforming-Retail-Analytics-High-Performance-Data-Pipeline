```python
import pandas as pd
from sqlalchemy import create_engine
import urllib.parse

# Correct credentials
db_user = 'postgres'
db_password = urllib.parse.quote_plus('#######')  # URL encode the password
db_host = 'retailpipeline.postgres.database.azure.com'
db_port = '5432'
db_name = 'retail_db'

# Enforce SSL mode for Azure PostgreSQL
connection_string = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}?sslmode=require'

# Create SQLAlchemy engine
engine = create_engine(connection_string)

# Test connection
try:
    with engine.connect() as conn:
        print("✅ Connected to PostgreSQL successfully!")
except Exception as e:
    print(f"❌ Connection failed: {e}")

# Insert DataFrames if they exist
try:
    if 'customer_df' in globals():
        customer_df.to_sql('customers', engine, if_exists='append', index=False)
    if 'products_df' in globals():
        products_df.to_sql('products', engine, if_exists='append', index=False)
    if 'stores_df' in globals():
        stores_df.to_sql('stores', engine, if_exists='append', index=False)
    print("✅ Data inserted successfully!")
except Exception as e:
    print(f"❌ Error inserting data: {e}")
```
